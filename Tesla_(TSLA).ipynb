```python
import pandas as pd
import matplotlib.pyplot as plt

# Данные
data = {
    'Date': ['2023-01-03', '2023-01-04', '2023-01-05', '2023-01-06', '2023-01-09'],
    'Close': [108.10, 113.64, 110.34, 113.06, 119.77],
    'High': [118.80, 114.59, 111.75, 114.39, 123.52],
    'Low': [104.64, 107.52, 107.16, 101.81, 117.11],
    'Open': [118.47, 109.11, 110.51, 103.00, 118.96],
}

# Создание DataFrame
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])  # Преобразование даты

# Построение графика
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Close'], marker='o', linestyle='-', label='Close Price')
plt.plot(df['Date'], df['High'], marker='o', linestyle='--', label='High Price')
plt.plot(df['Date'], df['Low'], marker='o', linestyle='--', label='Low Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('Tesla (TSLA) Stock Prices')
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.show()
```
